```python
from datetime import datetime
import sys
import os
from time import struct_time

# class for book. It store the title of the book, author, isbn and the quantity.
class Book:
    def __init__(self, title, author, isbn, quantity):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.quantity = quantity
        if quantity>0:
            available1=True
        else:
            available1=False
        self.available = available1   # added new field to track availability
 
class Library:
    def __init__(self):
        self.books = []
        self.borrowed_books = []
    
    def add_book(self, title, author, isbn, quantity):
        book = Book(title, author, isbn, quantity)
        self.books.append(book)
        print(f"Book '{book.title}' added to the library.")

    def display_books(self, index=0):
        i=1
        if index == 0:
            print("__________________________________________________________________________________")
            print("|                                                                                 |")
            print("|                        Books Available in the Library:                          |")
            print("|_________________________________________________________________________________|\n")
        if index < len(self.books):
            book = self.books[index]
            print ("===================================================================================")
            if book.available:
                i = index+1
                print(f"{i}- {book.title} by {book.author} \t \n (ISBN: {book.isbn}, Quantity: {book.quantity})\n")
                
            self.display_books(index+1)
        
                
    def borrow_book(self, isbn, student_id, student_name):
        for book in self.books:
            if book.isbn == isbn:
                if book.available and book.quantity>0:
                    book.quantity -=1  # mark book as borrowed
                    self.borrowed_books.append((book, student_id, student_name, datetime.now()))
                    print(f"Book '{book.title}' borrowed by {student_name}.")
                else:
                    print(f"Book '{book.title}' is not available for borrowing.")
                break
        else:
            print("Book not found in the library.")
            
    def return_book(self, isbn):
        for book, student_id, student_name, borrow_date in self.borrowed_books:
            if book.isbn == isbn:
                book.quantity +=1  # mark book as returned
                self.borrowed_books.remove((book, student_id, student_name, borrow_date))
                days_borrowed = (datetime.now() - borrow_date).days
                if days_borrowed > 7:
                    late_fee = (days_borrowed - 7) * 0.5
                    print(f"Book '{book.title}' returned by {student_name}. Late fee: ${late_fee:.2f}")
                else:
                    print(f"Book '{book.title}' returned by {student_name}. {days_borrowed}")
                break
        else:
            print("Book not found in the borrowed books list.")


library = Library()

library.add_book("Harry Potter and the Philosopher's Stone", "J.K. Rowling", "9780747532743", 5)
library.add_book("To Kill a Mockingbird", "Harper Lee", "9780446310789", 1)
library.add_book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565", 7)
library.add_book("Harry Potter and the Philosopher's Stone", "J.K. Rowling", "9780747532743", 5)
library.add_book("To Kill a Mockingbird", "Harper Lee", "9780446310789", 1)
library.add_book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565", 7)
library.add_book("1984", "George Orwell", "9780451524935", 2)
library.add_book("The Catcher in the Rye", "J.D. Salinger", "9780316769488", 2)
library.add_book("The Hobbit", "J.R.R. Tolkien", "9780547928227)", 3)
library.add_book("The Lord of the Rings", "J.R.R. Tolkien", "9780544003415", 2)
library.add_book("Animal Farm", "George Orwell", "9780451526342", 1)
library.add_book("The Diary of a Young Girl", "Anne Frank", "9780553296983", 4)
library.add_book("One Hundred Years of Solitude", "Gabriel Garcia", "9780060883287", 2)
library.add_book("The Picture of Dorian Gray", "Oscar Wilde", "9780141442464", 3)
library.add_book("The Adventures of Huckleberry Finn", "Mark Twain", "9780486280615", 1)
library.add_book("The Adventures of Tom Sawyer", "Mark Twain", "9780486400778", 1)
library.add_book("The Stranger", "Albert Camus", "9780679720201", 4)
library.add_book("The Sun Also Rises", "Ernest Hemingway", "9780684800714", 6)
library.add_book("Wuthering Heights", "Emily Bronte", "9780141439556", 5)
library.add_book("Jane Eyre", "Charlotte Bronte", "9780141441146", 3)
library.add_book("The Alchemist", "Paulo Coelho", "9780062315007", 2)
library.add_book("Bumi", "Tere Liye", "9786020313556", 6)
library.add_book("Laskar Pelangi", "Andrea Hirata", "9789799106680", 8)
library.add_book("Laskar Pelangi: Edensor", "Andrea Hirata", "9789799106680", 5)
library.add_book("Laskar Pelangi: Sang Pemimpi", "Andrea Hirata", "9789799106680", 6)
library.add_book("Bumi Manusia", "Pramoedya Ananta Toer", "9979278278603", 4)
library.add_book("Kimia untuk SMA Kelas 10", "Yusuf Kurniawan, dkk.", "9789790185066", 8)
library.add_book("Biologi SMA Kelas 11", "Asep Komarudin, dkk.", "9786022460814", 8)
library.add_book("Sejarah Indonesia Kelas 12", "Supriyono, dkk.", "9789794618817", 8)
library.add_book("Matematika SMA Kelas 12", "Wiwik Widayati, dkk.", "9789794618909", 8)
library.add_book("Bahasa Inggris untuk SMA Kelas 12", "M. Saleh Lubis, dkk.", "9789794618923", 8)
os.system('cls' if os.name == 'nt' else 'clear')

print(" ___________________________________________")
print("\n| Welcome to the Library Management System! |")
print(" ___________________________________________")

while True:

    print(" ___________________________________________")
    print("| Please choose an option:                  |")
    print("| 1. Add book                               |")
    print("| 2. Display books                          |")
    print("| 3. Borrow book                            |")
    print("| 4. Return book                            |")
    print("| 5. Track books                            |")
    print("| 6. Quit                                   |")
    print("|___________________________________________|")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        os.system('cls' if os.name == 'nt' else 'clear')

        title =  input("Enter the book title: ")
        author = input("Enter the book author: ")
        isbn = input("Enter the book ISBN: ")
        quantity = int(input("Enter the book quantity: "))
        library.add_book(title, author, isbn, quantity)
        
        
    elif choice == "2":
        os.system('cls' if os.name == 'nt' else 'clear')
        library.display_books()
        print("\n")


    elif choice == "3":
        os.system('cls' if os.name == 'nt' else 'clear')
        isbn = input("Enter the book ISBN: ")
        student_id = input("Enter your student ID: ")
        student_name = input("Enter your name: ")
        library.borrow_book(isbn, student_id, student_name)
        
    elif choice == "4":
        os.system('cls' if os.name == 'nt' else 'clear')
        isbn = input("Enter the book ISBN: ")
        library.return_book(isbn)
        
    elif choice == "5":
        os.system('cls' if os.name == 'nt' else 'clear')
        print (library.borrowed_books)

        for book in library.books:
            print(f"{book.title} by {book.author} (ISBN: {book.isbn}, Quantity: {book.quantity})")
            
    elif choice == "6":
        os.system('cls' if os.name == 'nt' else 'clear')
        print("Thank you for using the Library Management System!")
        break
        
    else:
        print("Invalid choice. Please try again.")
```
